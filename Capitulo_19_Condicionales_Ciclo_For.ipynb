## Estructuras Condicionales

Las **estructuras condicionales** permiten que el programa tome decisiones basadas en condiciones, igual que los historiadores evalÃºan evidencias antes de llegar a conclusiones.

```{admonition} ðŸ›ï¸ AnalogÃ­a HistÃ³rica
:class: note
Imagina que eres un historiador evaluando si un evento es anterior o posterior a cierta fecha. Si el aÃ±o es menor a 1810, entonces es perÃ­odo colonial; si es igual a 1810, es el inicio de la Independencia; de lo contrario, es perÃ­odo republicano.
```

### La estructura `if`

````python
# Clasificar un aÃ±o histÃ³rico
anio = 1818

if anio >= 1810:
    print(f"El aÃ±o {anio} pertenece al perÃ­odo de Independencia o posterior")
````

### La estructura `if-else`

````python
# Â¿Es perÃ­odo colonial o independiente?
anio = 1750

if anio < 1810:
    print(f"El aÃ±o {anio} pertenece al perÃ­odo Colonial")
else:
    print(f"El aÃ±o {anio} pertenece al perÃ­odo de Independencia o posterior")
````

### La estructura `if-elif-else`

Para mÃºltiples condiciones usamos `elif` (abreviatura de "else if"):

````python
# Clasificar perÃ­odos de la historia de Chile
anio = 1830

if anio < 1541:
    periodo = "Precolombino"
elif anio < 1810:
    periodo = "Colonial"
elif anio < 1818:
    periodo = "Patria Vieja / Reconquista"
elif anio < 1830:
    periodo = "Patria Nueva / OrganizaciÃ³n"
elif anio < 1891:
    periodo = "RepÃºblica Conservadora y Liberal"
else:
    periodo = "Chile ContemporÃ¡neo"

print(f"El aÃ±o {anio} corresponde al perÃ­odo: {periodo}")
````

## Operadores de ComparaciÃ³n

| Operador | Significado | Ejemplo | Resultado |
|----------|-------------|---------|----------|
| `==` | Igual a | `5 == 5` | `True` |
| `!=` | Diferente de | `5 != 3` | `True` |
| `<` | Menor que | `3 < 5` | `True` |
| `>` | Mayor que | `5 > 3` | `True` |
| `<=` | Menor o igual | `5 <= 5` | `True` |
| `>=` | Mayor o igual | `5 >= 3` | `True` |

*Tabla 17.1: Operadores de comparaciÃ³n en Python*

````python
# Ejemplos con contexto histÃ³rico
anio_rancagua = 1814  # Desastre de Rancagua
anio_maipu = 1818     # Batalla de MaipÃº

print(f"Â¿Rancagua fue antes que MaipÃº?: {anio_rancagua < anio_maipu}")
print(f"Â¿Son el mismo aÃ±o?: {anio_rancagua == anio_maipu}")
print(f"Â¿Rancagua fue en 1814?: {anio_rancagua == 1814}")
````

## Operadores LÃ³gicos

| Operador | Significado | Ejemplo |
|----------|-------------|----------|
| `and` | Y (ambas verdaderas) | `x > 0 and x < 10` |
| `or` | O (al menos una verdadera) | `x < 0 or x > 100` |
| `not` | NegaciÃ³n | `not (x > 5)` |

*Tabla 17.2: Operadores lÃ³gicos en Python*

````python
# Verificar si un aÃ±o estÃ¡ en el siglo XIX
anio = 1879  # Guerra del PacÃ­fico

# Un aÃ±o estÃ¡ en el siglo XIX si es >= 1801 Y <= 1900
en_siglo_xix = anio >= 1801 and anio <= 1900
print(f"Â¿{anio} estÃ¡ en el siglo XIX?: {en_siglo_xix}")

# Â¿Es un aÃ±o de guerra importante?
es_guerra_pacifico = anio >= 1879 and anio <= 1884
es_guerra_civil = anio == 1891
en_conflicto = es_guerra_pacifico or es_guerra_civil

print(f"Â¿Chile estaba en conflicto en {anio}?: {en_conflicto}")
````

## El Ciclo `for`

El ciclo `for` permite **iterar** (repetir) sobre una secuencia de elementos. Es como revisar uno por uno los documentos de un archivo histÃ³rico.

```{admonition} ðŸ“‹ Sintaxis bÃ¡sica
:class: tip
```python
for elemento in secuencia:
    # cÃ³digo a repetir
```
```

````python
# Iterar sobre una lista de estaciones del aÃ±o
estaciones = ["Primavera", "Verano", "OtoÃ±o", "Invierno"]

print("Las estaciones del aÃ±o son:")
for estacion in estaciones:
    print(f"  - {estacion}")
````

````python
# Iterar sobre presidentes chilenos del siglo XX
presidentes = [
    ("Arturo Alessandri", 1920),
    ("Carlos IbÃ¡Ã±ez", 1927),
    ("Pedro Aguirre Cerda", 1938),
    ("Salvador Allende", 1970)
]

print("Algunos presidentes del siglo XX:")
for nombre, anio in presidentes:
    print(f"  {nombre} asumiÃ³ en {anio}")
````

### La funciÃ³n `range()`

La funciÃ³n `range()` genera secuencias de nÃºmeros:

| Uso | Resultado |
|-----|------------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(2, 6)` | 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 |

*Tabla 17.3: Usos de la funciÃ³n range()*

````python
# Contar del 1 al 5
print("Contando del 1 al 5:")
for i in range(1, 6):
    print(i)

print("\nDÃ©cadas del siglo XX (solo cada 10 aÃ±os):")
for anio in range(1900, 2001, 10):
    print(f"DÃ©cada de {anio}")
````

### Iterar sobre strings

````python
# Contar letras en "CHILE"
palabra = "CHILE"

print(f"Letras en '{palabra}':")
for letra in palabra:
    print(f"  {letra}")

print(f"\nTotal de letras: {len(palabra)}")
````

### Combinando `for` con condicionales

````python
# Filtrar aÃ±os de guerra en Chile
anios = [1814, 1817, 1818, 1836, 1879, 1880, 1891, 1920, 1973]

print("AÃ±os de conflictos importantes:")
for anio in anios:
    # Guerra de Independencia (1810-1826)
    if 1810 <= anio <= 1826:
        print(f"  {anio}: Guerra de Independencia")
    # Guerra del PacÃ­fico (1879-1884)
    elif 1879 <= anio <= 1884:
        print(f"  {anio}: Guerra del PacÃ­fico")
    # Guerra Civil (1891)
    elif anio == 1891:
        print(f"  {anio}: Guerra Civil")
````

### Ciclos anidados: Listas dentro de listas

````python
# Series de TV organizadas por gÃ©nero
series_por_genero = [
    ["Los Simpson", "Friends", "The Office"],      # Comedia
    ["Game of Thrones", "Breaking Bad", "The Wire"]  # Drama
]

generos = ["Comedia", "Drama"]

for i, genero in enumerate(generos):
    print(f"\n{genero}:")
    for serie in series_por_genero[i]:
        print(f"  - {serie} ({len(serie)} caracteres)")
````

## ðŸ›ï¸ Ejemplo HistÃ³rico: AnÃ¡lisis de Batallas de la Independencia

````python
# Base de datos de batallas de la Independencia de Chile
batallas = [
    {"nombre": "Rancagua", "anio": 1814, "resultado": "Derrota"},
    {"nombre": "Chacabuco", "anio": 1817, "resultado": "Victoria"},
    {"nombre": "Cancha Rayada", "anio": 1818, "resultado": "Derrota"},
    {"nombre": "MaipÃº", "anio": 1818, "resultado": "Victoria"}
]

# AnÃ¡lisis
victorias = 0
derrotas = 0

print("="*50)
print("BATALLAS DE LA INDEPENDENCIA DE CHILE")
print("="*50)

for batalla in batallas:
    # Emoji segÃºn resultado
    emoji = "âœ…" if batalla["resultado"] == "Victoria" else "âŒ"
    
    print(f"{emoji} {batalla['nombre']} ({batalla['anio']}): {batalla['resultado']}")
    
    # Contar resultados
    if batalla["resultado"] == "Victoria":
        victorias += 1
    else:
        derrotas += 1

print("="*50)
print(f"RESUMEN: {victorias} victorias, {derrotas} derrotas")
print(f"Tasa de Ã©xito: {(victorias / len(batallas)) * 100:.1f}%")
````
