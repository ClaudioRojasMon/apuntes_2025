## Operaciones Matem√°ticas: Calculando la Historia

Python puede realizar operaciones matem√°ticas directamente. Esto es muy √∫til para calcular duraciones, poblaciones, distancias y otros datos hist√≥ricos.

### Operadores aritm√©ticos b√°sicos

| Operador | Operaci√≥n | Ejemplo | Resultado |
|:--------:|:----------|:--------|:---------:|
| `+` | Suma | `5 + 3` | `8` |
| `-` | Resta | `10 - 4` | `6` |
| `*` | Multiplicaci√≥n | `6 * 7` | `42` |
| `/` | Divisi√≥n | `20 / 4` | `5.0` |
| `**` | Potencia | `2 ** 3` | `8` |
| `//` | Divisi√≥n entera | `17 // 5` | `3` |
| `%` | M√≥dulo (resto) | `17 % 5` | `2` |

*Tabla 14.2: Operadores aritm√©ticos en Python*

````python
# ¬øCu√°ntos a√±os han pasado desde la Independencia de Chile?
a√±o_actual = 2025
a√±o_independencia = 1818

a√±os_transcurridos = a√±o_actual - a√±o_independencia
print(f"Han pasado {a√±os_transcurridos} a√±os desde la Independencia de Chile")
````

````python
# Calculemos la duraci√≥n del Imperio Romano de Occidente
# Desde la fundaci√≥n tradicional de Roma (753 a.C.) hasta la ca√≠da (476 d.C.)

inicio_roma = -753  # Usamos n√∫meros negativos para a.C.
caida_occidente = 476

duracion = caida_occidente - inicio_roma
print(f"El Imperio Romano (occidental) dur√≥ aproximadamente {duracion} a√±os")
````

````python
# Estimaci√≥n de poblaci√≥n: Si el Imperio Romano ten√≠a ~60 millones de habitantes
# y el mundo actual tiene ~8000 millones, ¬øcu√°ntas veces mayor es la poblaci√≥n actual?

poblacion_romana = 60_000_000  # Podemos usar guiones bajos para legibilidad
poblacion_actual = 8_000_000_000

proporcion = poblacion_actual / poblacion_romana
print(f"La poblaci√≥n mundial actual es {proporcion:.1f} veces mayor que la del Imperio Romano")
````

## Tipos de Datos B√°sicos

En Python, cada dato tiene un **tipo** que determina qu√© operaciones se pueden realizar con √©l. Los cuatro tipos b√°sicos son:

```{admonition} Los cuatro tipos b√°sicos
:class: important

1. **`int`** (entero): N√∫meros sin decimales ‚Üí `1810`, `-753`, `42`
2. **`float`** (flotante): N√∫meros con decimales ‚Üí `3.14`, `19.99`, `-273.15`
3. **`str`** (string/cadena): Texto entre comillas ‚Üí `"Bernardo O'Higgins"`, `'Chile'`
4. **`bool`** (booleano): Valores l√≥gicos ‚Üí `True`, `False`
```

````python
# Ejemplos con datos hist√≥ricos chilenos

# Entero (int): A√±o de un evento
a√±o_batalla_maipu = 1818
print(f"A√±o de la Batalla de Maip√∫: {a√±o_batalla_maipu}")
print(f"Tipo: {type(a√±o_batalla_maipu)}")
````

````python
# Flotante (float): Coordenadas geogr√°ficas
latitud_santiago = -33.45
longitud_santiago = -70.67
print(f"Coordenadas de Santiago: ({latitud_santiago}, {longitud_santiago})")
print(f"Tipo: {type(latitud_santiago)}")
````

````python
# String (str): Nombres y textos
procer_chileno = "Bernardo O'Higgins Riquelme"
cargo = 'Director Supremo de Chile'
print(f"{procer_chileno} fue {cargo}")
print(f"Tipo: {type(procer_chileno)}")
````

````python
# Booleano (bool): Verdadero o Falso
chile_es_republica = True
tiene_monarquia = False
print(f"¬øChile es una rep√∫blica? {chile_es_republica}")
print(f"¬øChile tiene monarqu√≠a? {tiene_monarquia}")
print(f"Tipo: {type(chile_es_republica)}")
````

### La funci√≥n `type()`

La funci√≥n `type()` nos permite verificar el tipo de cualquier dato:

````python
# Verificando tipos de datos
print(type(1810))           # int
print(type(3.14159))        # float
print(type("Historia"))     # str
print(type(True))           # bool
````

## Estructuras de Datos: Organizando la Historia

Adem√°s de los tipos b√°sicos, Python tiene **estructuras de datos** que permiten organizar m√∫ltiples valores.

### Listas (`list`)

Una **lista** es una colecci√≥n **ordenada** y **modificable** de elementos. Se crean usando corchetes `[]`.

```{admonition} Analog√≠a hist√≥rica
:class: tip
Piensa en una lista como un **cat√°logo de museo**: cada pieza tiene una posici√≥n espec√≠fica y podemos agregar o quitar piezas.
```

````python
# Lista de presidentes de Chile durante el siglo XX (algunos)
presidentes_chile = [
    "Arturo Alessandri Palma",
    "Carlos Ib√°√±ez del Campo", 
    "Pedro Aguirre Cerda",
    "Gabriel Gonz√°lez Videla",
    "Eduardo Frei Montalva",
    "Salvador Allende"
]

print("Algunos presidentes de Chile en el siglo XX:")
print(presidentes_chile)
````

````python
# Acceder a elementos por √≠ndice (¬°empieza en 0!)
print(f"Primer presidente en la lista: {presidentes_chile[0]}")
print(f"√öltimo presidente en la lista: {presidentes_chile[-1]}")
print(f"Total de presidentes en la lista: {len(presidentes_chile)}")
````

````python
# Las listas son modificables: podemos agregar elementos
presidentes_chile.append("Patricio Aylwin")
print(f"Lista actualizada: {presidentes_chile}")
````

### Tuplas (`tuple`)

Una **tupla** es una colecci√≥n **ordenada** pero **inmutable** (no se puede modificar). Se crean usando par√©ntesis `()`.

```{admonition} Analog√≠a hist√≥rica
:class: tip
Piensa en una tupla como una **inscripci√≥n en piedra**: una vez grabada, no puede modificarse.
```

````python
# Coordenadas geogr√°ficas de sitios hist√≥ricos (no deben cambiar)
coordenadas_palacio_moneda = (-33.4425, -70.6538)
print(f"Coordenadas del Palacio de La Moneda: {coordenadas_palacio_moneda}")
print(f"Tipo: {type(coordenadas_palacio_moneda)}")
````

### Diccionarios (`dict`)

Un **diccionario** es una colecci√≥n de pares **clave-valor**. Se crean usando llaves `{}`.

```{admonition} Analog√≠a hist√≥rica
:class: tip
Piensa en un diccionario como una **ficha de archivo hist√≥rico**: cada campo (clave) tiene un valor asociado.
```

````python
# Ficha de un personaje hist√≥rico
ficha_ohiggins = {
    "nombre": "Bernardo O'Higgins Riquelme",
    "nacimiento": 1778,
    "fallecimiento": 1842,
    "cargo": "Director Supremo de Chile",
    "periodo": "1817-1823",
    "lugar_nacimiento": "Chill√°n Viejo"
}

print("Ficha hist√≥rica de O'Higgins:")
for clave, valor in ficha_ohiggins.items():
    print(f"  {clave}: {valor}")
````

### Resumen: Comparaci√≥n de estructuras de datos

| Estructura | S√≠mbolo | Ordenada | Modificable | Ejemplo hist√≥rico |
|:-----------|:-------:|:--------:|:-----------:|:------------------|
| **Lista** | `[ ]` | ‚úÖ S√≠ | ‚úÖ S√≠ | Lista de batallas |
| **Tupla** | `( )` | ‚úÖ S√≠ | ‚ùå No | Coordenadas geogr√°ficas |
| **Diccionario** | `{ }` | ‚úÖ S√≠ | ‚úÖ S√≠ | Ficha biogr√°fica |
