# Prompting Efectivo: El arte de "conversar" con la IA

## ¿Qué es un "prompt"? <br>

- Un prompt es la instrucción o pregunta que le das a la IA. Es como darle órdenes a un asistente muy inteligente pero literal.
- Analogía: Imagina que tienes un asistente que sabe muchísimo pero necesita instrucciones muy claras. Si le dices "ayúdame con historia", no sabrá qué hacer. Pero si le dices "explícame las causas de la Independencia de Chile en 3 párrafos", tendrá claridad de como ayudarte. 

## Anatomía de un Prompt Efectivo <br>

**Componentes básicos**:

- Contexto: ¿Quién eres tú?
- Tarea: ¿Qué quieres que haga?
- Formato: ¿Cómo quieres la respuesta?
- Tono: ¿Cómo debe comunicarse?
- Restricciones: ¿Qué NO debe hacer?


### Prompts MAL Escritos vs Prompts Bien Escritos

**Ejemplo 1: Pedir explicaciones**

❌ MALO:"Háblame de la Independencia"<br>
Problema: Demasiado vago. La IA no sabe qué aspecto te interesa y de que independencia se trata.

✅ BUENO:
"Soy estudiante de 2° medio. Explícame en 3 párrafos cortos las causas principales de la Independencia de Chile. Usa un lenguaje simple y dame ejemplos concretos.

- "¿Por qué funciona?:

✅ Define tu nivel (2° medio)<br>
✅ Especifica qué quieres (causas)<br>
✅ Indica formato (3 párrafos)<br>
✅ Pide lenguaje apropiado<br>

**Ejemplo 2: Analizar fuentes históricas**

❌ MALO: "Analiza esto: Discurso Discurso de Churchill que te adjunto"
Problema: Demasiado vago. No entrega información desde que punto de vista quiere que analice el discurso:literario,Histórico, político, etc. 

✅ BUENO:
"Soy estudiante analizando fuentes primarias  de la Guerra del Peloponeso entre Esparata y Atenas . Este es un fragmento de la "Oracion Fúnebre" de Pericles:

Ayúdame a identificar:
1. El contexto histórico (qué estaba pasando en ese momento)
2. Lenguaje usado (formal/informal, emotivo/racional)
3. Quién era la audiencia
4. Qué mensaje principal quería transmitir"

**Ejemplo 3: Preparar presentación**

❌ MALO:"Dame ideas para mi presentación de Jose Manuel Balmaceda"

✅ BUENO:
"Debo hacer una presentación de 10 minutos sobre la vida de Jose Manuel Balmaceda - Presidente de Chile-  para mi clase de Historia PAES. Quiero enfocarme en aspectos menos  conocidos de su vida. 

Dame:
1. 3 aspectos poco conocidos pero interesantes.
2. Ideas de cómo presentar cada uno.
3. Sugerencias de imágenes o recursos visuales".

### Técnicas Avanzadas de Prompting

**Técnica 1: Rol Playing (Juego de roles)**

Pídele a la IA que asuma un personaje histórico: "Actúa como Bernardo O'Higgins en 1817, justo antes de la  Batalla de Chacabuco. Responde estas preguntas desde su perspectiva:

1. ¿Por qué es importante ganar esta batalla? <br>
2. ¿Qué te preocupa más?<br>
3. ¿Qué le dirías a tus soldados antes de la batalla?"<br>

✅  Útil para: entender motivaciones históricas, preparar debates, proyectos creativos

**Técnica 2: Comparación Estructurada**

"Compara la Independencia de Chile con la de Argentina usando la estructura siguiente:

| Aspecto | Chile | Argentina |
|---------|-------|-----------|
| Años del proceso | | |
| Líderes principales | | |
| Batallas clave | | |
| Apoyo externo | | |
| Resultado final | | |

Llena la tabla y luego explica las 3 diferencias más importantes"

**Técnica 3: Método Socrático (preguntas que llevan a más preguntas)**

"Estoy investigando sobre la Guerra del Pacífico. No me des respuestas directas, sino hazme 5 preguntas que me ayuden a entender mejor el tema y a descubrir aspectos importantes por mí mismo". 

✅ Útil para: desarrollar pensamiento crítico, preparar investigaciones.

**Técnica 4: Verificación Cruzada.**

"Dime que la Batalla de Maipú fue en 1818. ¿Es correcto?  Si sí, dame 3 fuentes donde pueda verificarlo. Si no, corrígeme y explica cuál es la fecha correcta"

**Técnica 5: Escalera de Complejidad,**

"Explícame el concepto de 'Patria Vieja' y 'Patria Nueva' en Chile en tres niveles:
.
- Nivel 1: Como si tuviera 10 años
- Nivel 2: Como si tuviera 15 años (mi nivel actual).
- Nivel 3: Como si fuera estudiante universitario". 

### Prompts Específicos para Tareas Comunes

**Para ensayos:**

"Voy a escribir un ensayo sobre XXX . Ayúdame a:

1. Generar 5 preguntas de investigación interesantes
2. Crear un esquema con introducción, 3 argumentos principales 
   y conclusión
3. Identificar posibles fuentes que debería consultar

NO escribas el ensayo, solo ayúdame a planificarlo"

**Para estudiar:**

    "Tengo prueba sobre la Guerra del Pacífico. Ayúdame a estudiar:

1. Hazme 10 preguntas de opción múltiple (4 alternativas cada una)
2. Dame las respuestas al final para que pueda corregirme
3. Para cada pregunta, explica por qué esa es la respuesta correcta"

**Para presentaciones:**

"Haré una presentación de 15 minutos sobre  XXX  para mi clase. 

Necesito:<br>
1. Un esquema de 5 diapositivas principales
2. Para cada diapositiva: título y 3 puntos clave
3. Sugerencia de qué imagen o gráfico poner
4. Un dato curioso o anécdota para captar la atención"

**Para análisis de textos:**

 "Lee este discurso de [personaje histórico] [pega el texto]

Analiza:<br>
1. Contexto histórico (¿cuándo y por qué se dio este discurso?)
2. Mensaje principal
3. Técnicas retóricas usadas
4. Impacto que pudo tener en su época"

### Ejemplos Prácticos con Personajes Históricos Chilenos

**Ejercicio 1: Entrevista a Arturo Prat**

Prompt: "Actúa como Arturo Prat el 20 de mayo de 1879, horas antes  del Combate Naval de Iquique. Responde esta entrevista  imaginaria desde su perspectiva:

1. ¿Cómo te sientes ante la batalla inminente?
2. ¿Por qué es importante defender Iquique?
3. ¿Qué mensaje le darías a tu familia?
4. ¿Qué significa para ti el honor y el deber?

Responde en primera persona, con el lenguaje y valores de  la época"

**Ejercicio 2: Debate entre O'Higgins y Carrera**

Prompt: "Simula un debate entre Bernardo O'Higgins y José Miguel Carrera  sobre la mejor estrategia para lograr la Independencia de Chile.

Formato: <br>
- 3 turnos para cada uno
- Cada turno: 1 párrafo con sus argumentos
- Al final, resume las diferencias principales entre ambas visiones

Usa lenguaje de la época pero comprensible para un estudiante actual"

**Ejercicio 3: Diario de Vida de Javiera Carrera**

Prompt: "Escribe 3 entradas de diario de Javiera Carrera durante la  Independencia de Chile:

Entrada 1: Cuando su hermano José Miguel asume el poder (1811)
Entrada 2: Durante el Desastre de Rancagua (1814)
Entrada 3: En el exilio en Argentina (1815)

Cada entrada debe:<br>
- Reflejar emociones personales
- Mencionar eventos históricos reales
- Mostrar el rol de las mujeres en la Independencia
- Usar lenguaje de la época pero accesible"

### ⚠️ Errores Comunes al Hacer Prompts

**Error 1: Ser demasiado vago** <br>
❌ "Ayúdame con historia"<br>
✅ "Explícame las causas de la Guerra Civil de 1891 en Chile"

**Error 2: No especificar el nivel**<br>
❌ "Explica la teoría de la relatividad".<br>
✅ "Explica la teoría de la relatividad como si tuviera 15 años"

**Error 3: Pedir que haga todo el trabajo**<br>
❌ "Escríbeme el ensayo completo".<br>
✅ "Ayúdame a crear un esquema para mi ensayo"

**Error 4: No verificar información**<br>
❌ Aceptar fechas sin verificar.<br>
✅ "Dame esta información y dime dónde puedo verificarla"

**Error 5: Prompts sin estructura.** <br>
❌ Un párrafo largo sin formato.<br>
✅ Usar números, bullets, secciones claras

### Actividad Práctica: Mejora estos prompts malos convirtiéndolos en prompts buenos

**Prompt Malo 1**: "Dime sobre Gabriela Mistral"

**Tu versión mejorada:** [Escribe aquí tu prompt mejorado]


**Prompt Malo 2:** "Hazme un resumen de la Guerra del Pacífico"

**Tu versión mejorada:** [Escribe aquí tu prompt mejorado]

**Prompt Malo 3:"Ayúdame con mi tarea de historia"**

**Tu versión mejorada:** [Escribe aquí tu prompt mejorado]

### Plantillas de Prompts Reutilizables

**Plantilla 1: Para entender conceptos**
"Soy estudiante de [nivel]. Explícame [concepto] en [cantidad] párrafos usando un lenguaje [simple/técnico]. Dame [número]  ejemplos concretos relacionados con [contexto específico]."



**Plantilla 2: Para analizar fuentes**:  "Analiza este [tipo de fuente] del año [año] sobre [tema]:[pega la fuente]

Identifica:<br>
1. [Aspecto 1]
2. [Aspecto 2]
3. [Aspecto 3]"

**Plantilla 3: Para comparar:** "Compara [elemento A] con [elemento B] considerando estos aspectos:

- [Aspecto 1]
- [Aspecto 2]
- [Aspecto 3]

Presenta la información en una tabla y luego explica las diferencias principales.
